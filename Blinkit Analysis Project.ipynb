## **DATA ANALYSIS PROJECT USING PYTHON - BLINKIT ANALYSYS**

### **KPI Requirements**
**Total Sales:** The overall revenue generated from all items sold.<br>
**Average Sales:** The average revenue per sale.<br>
**Number of Items:** The total count of different items sold.<br>
**Average Rating:** The average customer rating for items sold. 


### **Chart's Requirements**
**Total Sales by Fat Content:** Analyze the impact of fat content on total sales.<br>
**Total Sales by Item Type:**  Identify the performance of different item types in terms of total sales.<br>
**Fat Content by Outlet for Total Sales:** Compare total sales across different outlets segmented by fat content.<br>
**Total Sales by Outlet Establishment:** Evaluate how the age or type of outlet establishment influences total sales.<br>
**Sales by Outlet Size:** Analyze the correlation between outlet size and total sales.<br>
**Sales by Outlet Location:** Assess the geographic distribution of sales across different locations.

#### **Data Cleaning**

```python
print (df['Item Fat Content'].unique())  ## We have ['Regular' 'Low Fat' 'low fat' 'LF' 'reg']
```

```python
df['Item Fat Content'] = df['Item Fat Content'].replace({'LF':'Low Fat',
                                                         'low fat': 'Low Fat',
                                                         'reg':'Regular'})   ## replacing values
```

```python
print (df['Item Fat Content'].unique()) ## checking replaced values
```

### **BUSINESS REQUIREMENTS**

### **KPI's REQUIREMENTS**

```python
##Total Sales
total_sales= df['Sales'].sum()
## Avgerage Sales
avg_sales= df['Sales'].mean()
## No of items sold
no_of_items_sold= df['Sales'].count()
## Avgerage ratings
avg_ratings= df['Rating'].mean()

#Display

print(f"Total Sales: ${total_sales:,.1f}") ## f will help to concat a string with value and .1f says 1 decimal value 
print(f"Average Sales: ${avg_sales:,.1f}")
print(f"No Of Items Sold: {no_of_items_sold:,.1f}")
print(f"Average rating: {avg_ratings:,.1f}")
```

### **CHART's REQUIREMENTS**

#### **Total Sales By Fat Content**

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()
```

```python
plt.pie(sales_by_fat, labels = sales_by_fat.index,
        autopct = '%.0f%%', startangle = 90)
plt.title('Sales By fat Content')
plt.axis('equal')
plt.show()
```

```python
## Basic bar Graph

sales_by_type=df.groupby('Item Type')['Sales'].sum().sort_values (ascending = False)
bars =plt.bar(sales_by_type.index,sales_by_type.values)
plt.xticks(rotation=-90)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales By Item Type')
plt.tight_layout()
plt.show()
```

```python
## Bar Graph with data call-out values and resized graph

sales_by_type = df.groupby( 'Item Type') ['Sales'].sum().sort_values(ascending=False)
plt. figure(figsize=(10, 6))
bars = plt.bar(sales_by_type.index, sales_by_type.values)
plt.xticks(rotation=-90)
plt. xlabel( 'Item Type')
plt. ylabel('Total Sales')
plt. title('Total Sales by Item Type')
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             f'{bar.get_height():,.0f}', ha='center', va= 'bottom', fontsize=8)
plt. tight_layout()
plt. show()
```

```python
grouped = df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().unstack()
grouped = grouped[['Regular','Low Fat']]

ax = grouped.plot(kind='bar',figsize=(10,6))
plt.xticks(rotation =45)
plt.xlabel('Outlet L:ocation Tier')
plt.ylabel('Total Sales')
plt.title('Outlet Tier by Item Fat Content')
plt.legend(title='Item fat Content')
plt.tight_layout()
plt.show()
```

#### **Total Sales By Outlet Establishment**

```python
sales_by_year = df.groupby('Outlet Establishment Year')['Sales'].sum().sort_index()
plt.figure(figsize = (10,6))
plt.plot(sales_by_year.index, sales_by_year.values , marker = 's', linestyle= '--')
plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Total Sales by Outlet Establishment')

for x, y in zip(sales_by_year.index, sales_by_year.values+5):
    plt.text(x, y + 1000, f'{int(y):,}', ha='center', va='bottom', fontsize=9)
plt.show()
```

#### **Sales by Outlet Size**

```python
sales_by_size= df.groupby('Outlet Size')['Sales'].sum()

plt.pie(sales_by_size,labels =sales_by_size.index , autopct = '%1.0f%%')
plt.title('Sales by Outlet Size')
plt.tight_layout()
plt.show() 
```

#### **Sales by Outlet Location**

##### **Using MatPlotlib**

```python
sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum()
sales_by_location = sales_by_location.sort_values(ascending = True)
plt.figure(figsize=(8,5))
bars=plt.barh(sales_by_location.index , sales_by_location.values,color ='lightblue')
for bar in bars :
    width = bar.get_width()
    plt.text(width, bar.get_y() + bar.get_height() / 2
    ,f'{width:,.0f}',va='center', ha='left', color='black', fontweight='bold')
plt.xlabel('Total Sales')
plt.ylabel('Outlet Location Type')
plt.title('Total Sales By Outlet Location Type')  
plt.tight_layout()
plt.show()
```

##### **Using Seaborn**

```python
sales_by_location = df.groupby('Outlet Location Type')['Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values('Sales', ascending=False)
plt.figure(figsize=(8, 5)) # Smaller height, enough width
ax = sns.barplot(x='Sales', y='Outlet Location Type', data=sales_by_location)
plt.title('Total Sales by Outlet Location Type') 
plt.xlabel( 'Total Sales')
plt.ylabel( 'Outlet Location Type')
plt.tight_layout() # Ensures layout fits without scroll
plt.show()
```
